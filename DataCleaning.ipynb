```python
import pandas as pd
cchs11 = pd.read_csv("C:/CaseCompetition/Converted RData files/cchs11.csv")
cchs21 = pd.read_csv("C:/CaseCompetition/Converted RData files/cchs21.csv")
cchs31 = pd.read_csv("C:/CaseCompetition/Converted RData files/cchs31.csv")
```

```python
sub11=cchs11[['CCCA_121','CCCA_051','CCCA_05A','DHHAGAGE',
              'DHHA_SEX','DHHAGMS','SDCAGRAC','SDCAFIMM',
              'SDCAGRES','EDUADR04','INCAGHH','HWTAGBMI',
              'PACADPAI','TWDA_5','SMKADSTY','ALCADTYP',
             'CCCA_071','CCCA_101','CCCA_91B','FVCADTOT',
              'GENA_07','GEOAGPRV','WTSAM']]
sub21=cchs21[['CCCC_121','CCCC_051','CCCC_05A','DHHCGAGE',
              'DHHC_SEX','DHHCGMS','SDCCGRAC','SDCCFIMM',
              'SDCCGRES','EDUCDR04','INCCGHH','HWTCGBMI',
              'PACCDPAI','HCUC_1AA','SMKCDSTY','ALCCDTYP',
             'CCCC_071','CCCC_101','CCCC_91B','FVCCDTOT',
              'GENC_07','GEOCGPRV','WTSC_M']]

sub31=cchs31[['CCCE_121','CCCE_051','CCCE_05A','DHHEGAGE',
              'DHHE_SEX','DHHEGMS','SDCEGCGT','SDCEFIMM',
              'SDCEGRES','EDUEDR04','INCEGHH','HWTEGBMI',
              'PACEDPAI','HCUE_1AA','SMKEDSTY','ALCEDTYP',
             'CCCE_071','CCCE_101','CCCE_91F','FVCEDTOT',
              'GENE_07','GEOEGPRV','WTSE_M']]
```

```python
#各列Distinct值
for i in sub11.columns:
    print(i)
    print(pd.unique(sub11[i]))
    print('---------------------------------------------------')
```

```python
#选取Question 1 所需columns
#根据Note1删除缺失值
def CCHS_clean(dataset,columns,none_value):
    for i in columns:
        dataset = dataset[~dataset[i].isin(none_value)]
    return dataset

Note1=['NOT APPLICABLE','NOT STATED',"DON'T KNOW",'REFUSAL']
Q1=['CCCA_121','CCCA_05A','DHHAGAGE','DHHA_SEX', 'SDCAGRAC', 'EDUADR04',
         'INCAGHH', 'HWTAGBMI', 'TWDA_5', 'SMKADSTY', 'ALCADTYP',
         'CCCA_071', 'CCCA_101','PAC_05A']

Q2=['CCCA_121','CCCA_05A','DHHAGAGE','DHHA_SEX', 'SDCAGRAC', 'EDUADR04',
         'INCAGHH', 'HWTAGBMI', 'TWDA_5', 'SMKADSTY', 'ALCADTYP',
         'CCCA_071', 'CCCA_101','PAC_05A','SEX_05A','GMS_05A','RES_05A','PRV_05A']

#HouseHold Missing Value
#sub11.loc[sub11.INCAGHH == 'NOT STATED','INCAGHH'] = sub11.INCAGHH.mode()


Q_Note1=['CCCA_121','DHHAGMS','SDCAGRAC','EDUADR04','INCAGHH','HWTAGBMI','PACADPAI',
         'TWDA_5','SMKADSTY','ALCADTYP','CCCA_071','CCCA_101','CCCA_91B',
         'FVCADTOT','GENA_07','GEOAGPRV','FVCADTOT']
Q1_Note1=['CCCA_121','SDCAGRAC', 'EDUADR04', 'INCAGHH',
          'HWTAGBMI', 'TWDA_5', 'SMKADSTY', 'ALCADTYP',
         'CCCA_071', 'CCCA_101']
sub11 = CCHS_clean(sub11, Q_Note1, Note1)

#Recode  arthritis
sub11.loc[sub11.CCCA_05A.isin(['OSTEOARTHRITIS', 'OTHER', 'RHEUMATOID ARTH']),'CCCA_05A'] = 'YES'
sub11.loc[~sub11.CCCA_05A.isin(['YES']),'CCCA_05A']  = 'NO'

#Recode Province
sub11.loc[sub11.GEOAGPRV.isin(['YUKON/NWT/NUNAVT','YUKON/NWT/NUNA.']),'GEOAGPRV'] = 'NORTH'
sub11.loc[~sub11.GEOAGPRV.isin(['NORTH']),'GEOAGPRV']  = 'SOUTH'

#Recode BMI
sub11 = CCHS_clean(sub11, ['HWTAGBMI'], ['NOT APPLICABLE','NOT STATED'])
sub11.HWTAGBMI = pd.to_numeric(sub11['HWTAGBMI'])
c1=sub11.HWTAGBMI < 18.5
c2=sub11.HWTAGBMI > 25
sub11.loc[c1,'HWTAGBMI'] = '<18.5'
sub11.loc[c2,'HWTAGBMI'] = '>25'
sub11.loc[~sub11.HWTAGBMI.isin(['>25','<18.5']),'HWTAGBMI'] = '18.5~25'

#Recode Age
sub11 = sub11[~sub11.DHHAGAGE.isin(['12 TO 14 YEARS','15 TO 19 YEARS','65 TO 69 YEARS', '70 TO 74 YEARS', '75 TO 79 YEARS', '80 YEARS OR MORE'])]

#Recode Daily Consumption
sub11 = sub11.loc[~sub11.FVCADTOT.isin(['NOT STATED']),:]
sub11.FVCADTOT = pd.to_numeric(sub11['FVCADTOT'])
c1=sub11.FVCADTOT < 3
c2=sub11.FVCADTOT > 6
sub11.loc[c1,'FVCADTOT'] = '0~3'
sub11.loc[c2,'FVCADTOT'] = '>6'
sub11.loc[~sub11.FVCADTOT.isin(['0~3','>6']),'FVCADTOT'] = '3~6'

#X1X2
#南北
#性别
#Marital Status
#Length of time in Canada since immigration
sub11 = sub11.assign(PAC_05A = sub11.PACADPAI +sub11.CCCA_05A)
sub11 = sub11.assign(SEX_05A = sub11.DHHA_SEX +sub11.CCCA_05A)
sub11 = sub11.assign(GMS_05A = sub11.DHHAGMS +sub11.CCCA_05A)
sub11 = sub11.assign(RES_05A = sub11.SDCAGRES +sub11.CCCA_05A)
sub11 = sub11.assign(PRV_05A = sub11.GEOAGPRV +sub11.CCCA_05A)



#Recode Sampling Weight Master Weight
sub11.WTSAM = sub11.WTSAM/3


#删除alcohol中的异常值
sub11 = sub11[sub11.ALCADTYP.isin(['REGULAR DRINKER', 'FORMER DRINKER', 'OCC. DRINKER', 'NEVER DRANK'])]



"""
Has heart disease, Cultural / racial origin, Highest level of education - respondent,
Total household income from all sources, Body mass index, Has a regular medical doctor, 
Type of smoker, type of drinker, Has high blood pressure, Has diabetes
"""


#Those who answered ‘NO’ are considered as ‘NOT APPLICABLE’ in the next variable ‘kind of arthritis’. 
sub11[sub11.CCCA_051 == 'NO'].CCCA_05A = 'NOT APPLICABLE'
sub11[sub11.SDCAFIMM == 'NO'].SDCAGRES = 'NOT APPLICABLE'
sub11.loc[sub11.SDCAGRES == 'NOT APPLICABLE', 'SDCAGRES'] = 'NOT IMMIGRANT'
sub11 = CCHS_clean(sub11, ['CCCA_05A'], ["DON'T KNOW", 'REFUSAL', 'NOT STATED' ])


sub11 = sub11.drop(columns=['CCCA_051','SDCAFIMM'])
sub11_Q1 = sub11[Q1]
sub11_Q2 = sub11[Q2]

```

```python
#转换为Dummy variable
def add_dummy(dataset, columns):
    for i in columns:
        dataset_dummy = pd.get_dummies(dataset[i], prefix=i, drop_first=True)
        dataset = pd.concat([dataset, dataset_dummy], axis=1)
    return dataset

dummy_columns = ['CCCA_121', 'CCCA_05A', 'DHHAGAGE', 'DHHA_SEX', 'SDCAGRAC',
       'EDUADR04', 'INCAGHH', 'HWTAGBMI', 'TWDA_5', 'SMKADSTY', 'ALCADTYP',
       'CCCA_071', 'CCCA_101','PAC_05A']

sub11_Q1 = add_dummy(sub11_Q1, dummy_columns)
sub11_Q2 = add_dummy(sub11_Q2, Q2)
dummy_columns = ['CCCA_121','CCCA_05A','DHHAGAGE',
              'DHHA_SEX','DHHAGMS','SDCAGRAC',
              'SDCAGRES','EDUADR04','INCAGHH','HWTAGBMI',
              'PACADPAI','TWDA_5','SMKADSTY','ALCADTYP',
             'CCCA_071','CCCA_101','CCCA_91B','FVCADTOT',
              'GENA_07','GEOAGPRV','PAC_05A','SEX_05A','GMS_05A','RES_05A','PRV_05A']
sub11 = add_dummy(sub11, dummy_columns)
```
